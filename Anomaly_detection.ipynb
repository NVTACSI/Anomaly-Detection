Anomaly detection (AD) refers to the process of detecting data points that do not conform with the rest of observations. Applications of anomaly detection include fraud and fault detection, surveillance, diagnosis, data cleanup, predictive maintenance.

When we talk about AD, we usually look at it as an unsupervised (or semi-supervised) task, where the concept of anomaly is often not well defined or, in the best case, just few samples are labeled as anomalous. In this challenge, you will look at AD from a different perspective!

The dataset you are going to work on consists of monitoring data generated by IT systems; such data is then processed by a monitoring system that executes some checks and detects a series of anomalies. This is a multi-label classification problem, where each check is a binary label corresponding to a specific type of anomaly. Your goal is to develop a machine learning model (or multiple ones) to accurately detect such anomalies.

This will also involve a mixture of data exploration, pre-processing, model selection, and performance evaluation. You will also be asked to try one or more rule learning models, and compare them with other ML models both in terms of predictive performances and interpretability. Interpreatibility is indeed a strong requirement especially in applications like AD where understanding the output of a model is as important as the output itself.

## 1.3 Features Analysis

In order to have better insights as well as easily figure out outliers of features, we use a box plot for each feature. The box plot allows us to have a nice summary of numeric variables. The line that divides the box into 2 parts represents the median of the data. The end of the box shows the upper and lower quartiles. The extreme lines shows the highest and lowest value excluding outliers.

```python
def plot_box_distribution(feature):
    """Plot box plot for features.
    
    Parameters
    ----------
    feature: string
        Name of the plotted feature
    """
    # Visualize the box plot of features
    
    _, ax = plt.subplots(2, 4, figsize=(15, 8))

    # Plot the count plot
    for i in range(len(targets)):
        tempDF = trainDF.dropna(subset=[feature, targets[i]])
        tempDF[targets[i]] = tempDF[targets[i]].astype(int)
        sns.boxplot(x=targets[i], y=feature, hue=targets[i], data=tempDF,
                    ax=ax[i//4, i%4], linewidth=2.5, dodge=False)
        ax[i//4, i%4].legend(loc=0)
        ax[i//4, i%4].set_title(targets[i], weight="semibold")
        ax[i//4, i%4].set_ylabel("")
        ax[i//4, i%4].set_xlabel("")
```

Acutally, missing values is one of the most important aspects that we must care when doing data analysis and prediction. Detecting them is really easy (and this part is the proof) but the most difficult part comes when we have to fill it. How do we decide the right value? It is a crucial decision because a wrong or misleading value leads to bad predictions.

### 2.1.1 Explainatory attributes

```python
# Get the list of columns having missing value
missingAttrs = featuresDF.columns[featuresDF.isnull().any()].tolist()
                                         
                                         
# Count the missing values and remove the columns without missing values
missingCounts = featuresDF.isnull().sum()
missingCounts = missingCounts[missingCounts > 0]

# Sort by the number of missing values
missingCounts.sort_values(inplace=True)

# Build a data frame containing the information of missing data
missingDF = pd.DataFrame(list(zip(missingCounts.index.values, missingCounts.values)),
                         columns=['Feature', 'MissingValues'])
```

We can see there are lots of features which have missing value with different ratios. Since we do not have much domain knowledge on the problem, at this stage we temporarily remove samples which have missing value out of dataset and see the performance of models. If the models still work well, we will ignore these samples. But if not, we will consider some methods to handle missing value cases.

```python
## Remove samples which has missing values
attrs = list(set(trainDF.columns) - set(targets))

tempDF = trainDF.dropna(subset=attrs)
targetTrainDF = pd.DataFrame(tempDF[targets])
featuresTrainDF = tempDF.drop(columns=targets + ['SessionNumber', 'Date'])

tempDF = valDF.dropna(subset=attrs)
targetValDF = pd.DataFrame(tempDF[targets])
featuresValDF = tempDF.drop(columns=targets + ['SessionNumber', 'Date'])

tempDF = testDF.dropna(subset=attrs)
targetTestDF = pd.DataFrame(tempDF[targets])
featuresTestDF = tempDF.drop(columns=targets + ['SessionNumber', 'Date'])
```

### 2.1.2 Target attributes

The given dataset contains 8 different anomaly types and each type have binary labels. The null label's value on each type is in different samples. Therefore, we remove null label samples responding to each type.

```python
dictTrainDF = {'features' : [], 'target' : []}
dictValDF = {'features' : [], 'target' : []}
dictTestDF = {'features' : [], 'target' : []}

for target in targets:
    
    nullIdx = targetTrainDF[target].isnull()
    dictTrainDF['features'].append(featuresTrainDF[~nullIdx])
    dictTrainDF['target'].append(targetTrainDF[~nullIdx][target])
    
    nullIdx = targetValDF[target].isnull()
    dictValDF['features'].append(featuresValDF[~nullIdx])
    dictValDF['target'].append(targetValDF[~nullIdx][target])
    
    nullIdx = targetTestDF[target].isnull()
    dictTestDF['features'].append(featuresTestDF[~nullIdx])
    dictTestDF['target'].append(targetTestDF[~nullIdx][target])
```

# 3. Model Selection

Model Selection can be refered as selecting the best learning method (and their corresponding “optimal” hyperparameters) from a set of eligible machine learning methods. In this notebook, we apply `Decision tree` technique and see how it work on the given dataset. The reason we focus on this technique is it gives the interpretations for output of a model when important features will be on top of the tree and these interpretations is important in many field, especially in anomaly detection.

<p style='text-align: justify;'>
** 3.1 Decision tree **<br> 

<br> 
A decision tree is a flowchart-like structure in which each internal node represents a test on a feature (e.g. whether a coin flip comes up heads or tails) , each leaf node represents a class label (decision taken after computing all features) and branches represent conjunctions of features that lead to those class labels. The paths from root to leaf represent classification rules. While making decision tree, at each node of tree we ask different type of questions. Based on the asked question we will calculate the information gain corresponding to it. Algorithm for constructing decision tree usually works top-down, by choosing a variable at each step that best splits the set of items. Different algorithms use different metrices for measuring best.
<br> 
<br> 
The main parameters to adjust when using this methods is  `max_features`, which is the size of the random subsets of features to consider when splitting a node. The lower the greater the reduction of variance, but also the greater the increase in bias.
<br> 
<br>

<p align="center"> 
    <img src="https://static1.squarespace.com/static/59e19ba02278e7ca5b9ffb2b/t/5a07a3bf8165f51550358b35/1510450172549/decisiontree.png" alt="drawing" width="600">
</p>

As we already mentioned previously, we have 8 different anomaly types as known as 8 sub-task hence we build 8 binary model for them and stack into a dataframe for easily analysis later.

```python
modelDF = pd.DataFrame(columns=['model', 'TrainAcc', 'ValAcc'])

for i in range(len(targets)):
    Xtrain = dictTrainDF['features'][i]
    ytrain = dictTrainDF['target'][i]
    model = DecisionTreeClassifier(criterion='entropy', max_depth=15, min_samples_leaf=1)
    model.fit(Xtrain, ytrain)
    TrainAcc = f1_score(ytrain, model.predict(Xtrain), average='macro')
    
    Xval = dictValDF['features'][i]
    yval = dictValDF['target'][i]
    ValAcc = f1_score(yval, model.predict(Xval), average='macro')
    
    modelDF = modelDF.append({'model':model, 'TrainAcc':TrainAcc, 'ValAcc':ValAcc}, ignore_index=True)
```

Now, let's see the training and validation performance of our setting model

As we can see, Decision tree can learn patterns in the given dataset (even imbalance data) so that their prediction is highly accurate. However, accuracy is not enough for this type of problem (abnormally detection) when interpretations for output of a model is also crucial. Decision tree will put most important features on the top of tree and then we build graph and show top 3 branchs for each tree. Because we reuse these plots for categorical and oridinal features, we define a wrapper function for later reuse.

```python
def visualize_tree(tree):

    """Plot box plot and distribution plot for features.
    
    Parameters
    ----------
    tree: sckit-learn/Decision tree model.
        The machine learning model needed to be visualized
    
    Returns
    -------
    graph: the graph represent the Dot data (representation of the input tree).
    """
    
    dot_data = export_graphviz(tree, max_depth=3, out_file=None,  
                    filled=True, rounded=True,
                    special_characters=True)
    graph = pydotplus.graph_from_dot_data(dot_data)  
    
    return graph
```
